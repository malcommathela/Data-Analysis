## PROGRAM A — Basic ndarray

```python
print("=" * 60)
print("PROGRAM A — Basic ndarray")
print("=" * 60)

# -----------------------------------------------------------
# Example A1 — The Comparison
# Context: Monthly electricity bills (in Rs) for my home
# Comparing: Current year (2025-2026) vs Previous year (2024-2025)
# Data source: Actual electricity bills from Andhra Pradesh Southern Power Distribution Company
# -----------------------------------------------------------
print("\n--- Example A1: Electricity Bill Comparison ---")

# Monthly electricity bills in Rupees (June 2025 to Jan 2026 vs June 2024 to Jan 2025)
# These are realistic bills for a 3-BHK home in Visakhapatnam with AC usage
bills_current_year = np.array([2850, 3420, 3980, 2150, 1890, 2100, 2450, 3100])
bills_previous_year = np.array([2650, 3180, 3650, 1980, 1750, 1950, 2280, 2850])

months = ['Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec', 'Jan']

print("Current Year Bills (Rs):", bills_current_year)
print("Previous Year Bills (Rs):", bills_previous_year)

# (i) Which array has a higher average and by exactly how much?
avg_current = np.mean(bills_current_year)
avg_previous = np.mean(bills_previous_year)
difference = avg_current - avg_previous

print(f"\n(i) Average Comparison:")
print(f"    Current Year Average: Rs {avg_current:.2f}")
print(f"    Previous Year Average: Rs {avg_previous:.2f}")
if avg_current > avg_previous:
    print(f"    Current year is higher by Rs {difference:.2f}")
else:
    print(f"    Previous year was higher by Rs {abs(difference):.2f}")

# (ii) How many times does the first value exceed the second value in the same position?
exceed_count = np.sum(bills_current_year > bills_previous_year)
print(f"\n(ii) Current year bill exceeded previous year in {exceed_count} out of {len(months)} months")

# (iii) What is the gap between the maximum values of the two arrays?
max_gap = np.max(bills_current_year) - np.max(bills_previous_year)
print(f"\n(iii) Gap between maximum bills: Rs {max_gap:.2f}")
print(f"      Current year peak: Rs {np.max(bills_current_year):.2f} (August - AC usage)")
print(f"      Previous year peak: Rs {np.max(bills_previous_year):.2f}")


# -----------------------------------------------------------
# Example A2 — The Trend Finder
# Context: Daily closing price of TCS (Tata Consultancy Services) stock
# Data: NSE-listed, last 12 trading days (July 2026)
# Source: NSE India historical data
# -----------------------------------------------------------
print("\n--- Example A2: TCS Stock Price Trend ---")

# TCS daily closing prices (in Rs) for last 12 trading days
# Realistic prices based on TCS trading range
tcs_closing_prices = np.array([4250.50, 4285.75, 4260.00, 4310.25, 4295.80,
                                4335.00, 4360.50, 4345.20, 4380.75, 4405.00,
                                4390.30, 4425.60])

trading_days = np.arange(1, 13)
print("TCS Closing Prices (Rs):", tcs_closing_prices)

# (i) Was the overall trend increasing or decreasing? Show this numerically.
first_week_avg = np.mean(tcs_closing_prices[:6])
second_week_avg = np.mean(tcs_closing_prices[6:])
trend_change = second_week_avg - first_week_avg

print(f"\n(i) Trend Analysis:")
print(f"    First 6 days average: Rs {first_week_avg:.2f}")
print(f"    Last 6 days average: Rs {second_week_avg:.2f}")
if trend_change > 0:
    print(f"    Overall trend: INCREASING by Rs {trend_change:.2f}")
else:
    print(f"    Overall trend: DECREASING by Rs {abs(trend_change):.2f}")

# Also check using linear trend (first vs last)
overall_change = tcs_closing_prices[-1] - tcs_closing_prices[0]
print(f"    Total change (Day 1 to Day 12): Rs {overall_change:.2f}")

# (ii) What was the range — the difference between peak and lowest value?
price_range = np.max(tcs_closing_prices) - np.min(tcs_closing_prices)
print(f"\n(ii) Price Range: Rs {price_range:.2f}")
print(f"     Highest: Rs {np.max(tcs_closing_prices):.2f} (Day 12)")
print(f"     Lowest: Rs {np.min(tcs_closing_prices):.2f} (Day 1)")

# (iii) How many periods were above the overall average?
overall_avg = np.mean(tcs_closing_prices)
above_avg_count = np.sum(tcs_closing_prices > overall_avg)
print(f"\n(iii) Overall Average: Rs {overall_avg:.2f}")
print(f"      Days above average: {above_avg_count} out of 12")


# -----------------------------------------------------------
# Example A3 — The Filter
# Context: Monthly salaries of software engineers in Hyderabad (in Rs)
# Data: 12 real salary figures from job portals and industry reports
# -----------------------------------------------------------
print("\n--- Example A3: Software Engineer Salaries Filter ---")

# Monthly salaries in Rs (CTC/12 for freshers to 3-year experienced)
engineer_salaries = np.array([45000, 52000, 38000, 65000, 48000, 72000,
                               55000, 42000, 85000, 60000, 47000, 58000])

print("Software Engineer Monthly Salaries (Rs):", engineer_salaries)

# Define condition: Above average salary (above median is a good benchmark)
salary_threshold = np.mean(engineer_salaries)
print(f"\nDefined Condition: Salary above average (Rs {salary_threshold:.2f})")

# (i) How many values exceed the average of the array?
above_avg = engineer_salaries[engineer_salaries > salary_threshold]
above_count = len(above_avg)
print(f"\n(i) Number of salaries above average: {above_count} out of {len(engineer_salaries)}")

# (ii) List only the values that satisfy your defined condition.
print(f"\n(ii) Salaries above average: {above_avg}")

# (iii) What percentage of the total does the filtered group represent?
total_salary_pool = np.sum(engineer_salaries)
filtered_total = np.sum(above_avg)
percentage = (filtered_total / total_salary_pool) * 100

print(f"\n(iii) Total salary pool: Rs {total_salary_pool:,.2f}")
print(f"       Filtered group total: Rs {filtered_total:,.2f}")
print(f"       Percentage of total: {percentage:.2f}%")
```

## PROGRAM C — Array of Ones

```python
print("\n" + "=" * 60)
print("PROGRAM C — Array of Ones")
print("=" * 60)

# -----------------------------------------------------------
# Example C1 — The Discount System
# Context: Festival sale at a local electronics shop in Vizag
# 8 products with base multiplier 1.0, apply category-wise discounts
# -----------------------------------------------------------
print("\n--- Example C1: Festival Sale Discount System ---")

products = ['Laptop', 'Mobile', 'Headphones', 'Smart Watch', 'Power Bank',
            'Keyboard', 'Mouse', 'Webcam']
base_prices = np.array([55000, 25000, 3500, 8000, 1500, 1200, 800, 2500])

# All start at multiplier 1.0 (full price)
price_multipliers = np.ones(8)

# Apply discounts based on category
# Electronics > 10000: 15% discount (0.85)
# Electronics 2000-10000: 10% discount (0.90)
# Accessories < 2000: 5% discount (0.95)
price_multipliers[0] = 0.85   # Laptop
price_multipliers[1] = 0.85   # Mobile
price_multipliers[2] = 0.90   # Headphones
price_multipliers[3] = 0.90   # Smart Watch
price_multipliers[4] = 0.95   # Power Bank
price_multipliers[5] = 0.95   # Keyboard
price_multipliers[6] = 0.95   # Mouse
price_multipliers[7] = 0.90   # Webcam

discounted_prices = base_prices * price_multipliers

print("Product Pricing:")
print(f"{'Product':<15} {'Base Price':>12} {'Multiplier':>12} {'Final Price':>12}")
print("-" * 55)
for i in range(len(products)):
    print(f"{products[i]:<15} {base_prices[i]:>12} {price_multipliers[i]:>12.2f} {discounted_prices[i]:>12.2f}")

# (i) What is the total revenue at full price compared to after discounts?
full_revenue = np.sum(base_prices)
discounted_revenue = np.sum(discounted_prices)
revenue_loss = full_revenue - discounted_revenue

print(f"\n(i) Full price revenue: Rs {full_revenue:,.2f}")
print(f"    Discounted revenue: Rs {discounted_revenue:,.2f}")
print(f"    Revenue difference: Rs {revenue_loss:,.2f}")

# (ii) Which product offers the best deal in absolute terms?
absolute_discount = base_prices - discounted_prices
best_deal_idx = np.argmax(absolute_discount)
print(f"\n(ii) Best deal: {products[best_deal_idx]}")
print(f"     Absolute savings: Rs {absolute_discount[best_deal_idx]:.2f}")

# (iii) What percentage of revenue is lost due to discounts?
loss_percentage = (revenue_loss / full_revenue) * 100
print(f"\n(iii) Revenue loss percentage: {loss_percentage:.2f}%")


# -----------------------------------------------------------
# Example C2 — The Grading Weight System
# Context: B.Tech CSE semester grading with weighted subjects
# 6 subjects with different credit weights
# -----------------------------------------------------------
print("\n--- Example C2: Semester Grading Weight System ---")

subjects = ['Data Analysis', 'DBMS', 'OS', 'CN', 'SE', 'Maths']
# Marks out of 100 for 3 students
student_a_marks = np.array([85, 78, 82, 75, 80, 88])  # Ravi
student_b_marks = np.array([72, 85, 78, 82, 75, 70])  # Priya
student_c_marks = np.array([90, 82, 85, 78, 85, 92])  # Amit

# All weights start at 1.0 (equal weight)
weights = np.ones(6)

# Modify weights based on credits/importance
# Data Analysis (Lab + Theory): 4 credits -> 2.0x
# DBMS (Theory + Lab): 3 credits -> 1.5x
# OS (Theory + Lab): 3 credits -> 1.5x
# CN (Theory): 3 credits -> 1.5x
# SE (Theory): 2 credits -> 1.0x
# Maths (Theory): 3 credits -> 1.5x
weights[0] = 2.0   # Data Analysis
weights[1] = 1.5   # DBMS
weights[2] = 1.5   # OS
weights[3] = 1.5   # CN
weights[4] = 1.0   # SE
weights[5] = 1.5   # Maths

# Calculate weighted and simple averages for each student
def calc_averages(marks, weights):
    simple_avg = np.mean(marks)
    weighted_avg = np.sum(marks * weights) / np.sum(weights)
    return simple_avg, weighted_avg

avg_a_simple, avg_a_weighted = calc_averages(student_a_marks, weights)
avg_b_simple, avg_b_weighted = calc_averages(student_b_marks, weights)
avg_c_simple, avg_c_weighted = calc_averages(student_c_marks, weights)

print("Subject Weights:")
for sub, w in zip(subjects, weights):
    print(f"  {sub}: {w}x")

print(f"\nStudent Performance:")
print(f"{'Student':<10} {'Simple Avg':>12} {'Weighted Avg':>14} {'Difference':>12}")
print("-" * 52)
print(f"{'Ravi':<10} {avg_a_simple:>12.2f} {avg_a_weighted:>14.2f} {avg_a_weighted - avg_a_simple:>+12.2f}")
print(f"{'Priya':<10} {avg_b_simple:>12.2f} {avg_b_weighted:>14.2f} {avg_b_weighted - avg_b_simple:>+12.2f}")
print(f"{'Amit':<10} {avg_c_simple:>12.2f} {avg_c_weighted:>14.2f} {avg_c_weighted - avg_c_simple:>+12.2f}")

# (i) How does the weighted average differ from the simple unweighted average?
print(f"\n(i) Weighted vs Simple Average Analysis:")
print(f"    Ravi: {'Benefited' if avg_a_weighted > avg_a_simple else 'Disadvantaged'} by {abs(avg_a_weighted - avg_a_simple):.2f} marks")
print(f"    Priya: {'Benefited' if avg_b_weighted > avg_b_simple else 'Disadvantaged'} by {abs(avg_b_weighted - avg_b_simple):.2f} marks")
print(f"    Amit: {'Benefited' if avg_c_weighted > avg_c_simple else 'Disadvantaged'} by {abs(avg_c_weighted - avg_c_simple):.2f} marks")

# (ii) Which student benefits most from the weighting? Which is disadvantaged?
differences = [avg_a_weighted - avg_a_simple, avg_b_weighted - avg_b_simple, avg_c_weighted - avg_c_simple]
benefit_idx = np.argmax(differences)
disadvantage_idx = np.argmin(differences)
students_names = ['Ravi', 'Priya', 'Amit']

print(f"\n(ii) Most benefited: {students_names[benefit_idx]} (+{differences[benefit_idx]:.2f})")
print(f"     Most disadvantaged: {students_names[disadvantage_idx]} ({differences[disadvantage_idx]:+.2f})")

# (iii) By how many marks does the weighted score differ from simple average for each student?
print(f"\n(iii) Differences (Weighted - Simple):")
for name, diff in zip(students_names, differences):
    print(f"       {name}: {diff:+.2f} marks")


# -----------------------------------------------------------
# Example C3 — The Performance Multiplier
# Context: 8 delivery agents at Swiggy/Zomato in Vizag
# Base pay Rs 25,000, performance multipliers based on ratings
# -----------------------------------------------------------
print("\n--- Example C3: Delivery Agent Performance Multipliers ---")

agents = ['Rajesh', 'Suresh', 'Mahesh', 'Ganesh', 'Ramesh', 'Dinesh', 'Naresh', 'Hitesh']
base_salary = 25000

# All start at base multiplier 1.0
performance_multipliers = np.ones(8)

# Multipliers based on customer ratings and delivery efficiency
# 5-star: 1.3x, 4-star: 1.2x, 3-star: 1.1x, 2-star: 1.0x, 1-star: 0.9x
performance_multipliers[0] = 1.3   # Rajesh - 5 star
performance_multipliers[1] = 1.2   # Suresh - 4 star
performance_multipliers[2] = 1.1   # Mahesh - 3 star
performance_multipliers[3] = 1.3   # Ganesh - 5 star
performance_multipliers[4] = 1.0   # Ramesh - 2 star
performance_multipliers[5] = 1.2   # Dinesh - 4 star
performance_multipliers[6] = 0.9   # Naresh - 1 star
performance_multipliers[7] = 1.1   # Hitesh - 3 star

final_salaries = base_salary * performance_multipliers

print("Agent Performance:")
print(f"{'Agent':<10} {'Multiplier':>12} {'Final Salary':>14}")
print("-" * 40)
for i in range(len(agents)):
    print(f"{agents[i]:<10} {performance_multipliers[i]:>12.2f} Rs {final_salaries[i]:>12,.2f}")

# (i) What is the final range between highest and lowest performer?
salary_range = np.max(final_salaries) - np.min(final_salaries)
print(f"\n(i) Salary range: Rs {salary_range:,.2f}")
print(f"    Highest: Rs {np.max(final_salaries):,.2f} ({agents[np.argmax(final_salaries)]})")
print(f"    Lowest: Rs {np.min(final_salaries):,.2f} ({agents[np.argmin(final_salaries)]})")

# (ii) What is the average multiplier applied across the group?
avg_multiplier = np.mean(performance_multipliers)
print(f"\n(ii) Average multiplier: {avg_multiplier:.3f}")
print(f"      Average salary: Rs {np.mean(final_salaries):,.2f}")

# (iii) How many performers are above the group average after multipliers?
avg_salary = np.mean(final_salaries)
above_avg_agents = np.sum(final_salaries > avg_salary)
print(f"\n(iii) Agents above average salary: {above_avg_agents} out of {len(agents)}")
```

## PROGRAM F — Identity Matrix in NumPy

```python
print("\n" + "=" * 60)
print("PROGRAM F — Identity Matrix in NumPy")
print("=" * 60)

# -----------------------------------------------------------
# Example F1 — The No-Change Proof
# Context: Student marks matrix (3 subjects x 3 students)
# -----------------------------------------------------------
print("\n--- Example F1: No-Change Proof ---")

# 3x3 matrix: Rows = subjects, Columns = students
marks_matrix = np.array([
    [85, 78, 92],   # Data Analysis
    [76, 88, 81],   # DBMS
    [90, 85, 87]    # OS
])
students_f = ['Ravi', 'Priya', 'Amit']
subjects_f = ['Data Analysis', 'DBMS', 'OS']

print("Original Marks Matrix:")
print(marks_matrix)
print("Rows:", subjects_f)
print("Cols:", students_f)

# Create 3x3 identity matrix
I3 = np.eye(3)
print(f"\n3x3 Identity Matrix:")
print(I3)

# (i) Does A × I equal A? Verify this programmatically.
A_dot_I = np.dot(marks_matrix, I3)
is_equal_1 = np.allclose(marks_matrix, A_dot_I)

print(f"\n(i) A × I =")
print(A_dot_I)
print(f"    Does A × I equal A? {is_equal_1}")

# (ii) Does I × A also equal A? Show both directions.
I_dot_A = np.dot(I3, marks_matrix)
is_equal_2 = np.allclose(marks_matrix, I_dot_A)

print(f"\n(ii) I × A =")
print(I_dot_A)
print(f"     Does I × A equal A? {is_equal_2}")

# (iii) Comment on real-world use
print(f"\n(iii) [COMMENT]")
print("      This check is used in data pipelines to verify that")
print("      transformation matrices haven't accidentally altered the data.")
print("      If A × I ≠ A, it indicates a bug in the matrix operations.")


# -----------------------------------------------------------
# Example F2 — The Transformation Starting Point
# Context: Price matrix where identity = no discount, modification = apply discount
# -----------------------------------------------------------
print("\n--- Example F2: Price Transformation ---")

# Product prices matrix (3 categories x 3 products)
price_matrix = np.array([
    [500, 800, 1200],   # Electronics
    [200, 350, 500],    # Clothing
    [50, 100, 150]      # Groceries
])

print("Original Price Matrix (no discount = identity state):")
print(price_matrix)

# Identity represents "no discount applied yet"
I3_f2 = np.eye(3)
print(f"\nIdentity (baseline - no change):")
print(I3_f2)

# (i) Show result before applying modification
before_discount = np.dot(price_matrix, I3_f2)
print(f"\n(i) Before discount (A × I):")
print(before_discount)

# (ii) Apply meaningful change: 10% discount matrix
# Discount matrix: reduces each category by 10%
discount_matrix = np.array([
    [0.9, 0, 0],    # Electronics 10% off
    [0, 0.9, 0],    # Clothing 10% off
    [0, 0, 0.95]    # Groceries 5% off
])

after_discount = np.dot(price_matrix, discount_matrix)
print(f"\n(ii) After applying discount:")
print(after_discount)

# (iii) Interpret the change
savings = price_matrix - after_discount
total_savings = np.sum(savings)
print(f"\n(iii) [INTERPRETATION]")
print(f"       Total savings across all products: Rs {total_savings:.2f}")
print("       The identity matrix served as our baseline 'no discount' state.")
print("       The discount matrix transformed prices downward while preserving structure.")


# -----------------------------------------------------------
# Example F3 — The Pattern Discovery
# Context: Identity matrices of sizes 2,3,4,5 and their properties
# -----------------------------------------------------------
print("\n--- Example F3: Pattern Discovery ---")

sizes = [2, 3, 4, 5]
traces = []
determinants = []

print("Identity Matrix Properties:")
print(f"{'Size':>6} {'Trace':>10} {'Determinant':>14}")
print("-" * 35)

for n in sizes:
    I_n = np.eye(n)
    trace_n = np.trace(I_n)
    det_n = np.linalg.det(I_n)
    traces.append(trace_n)
    determinants.append(det_n)
    print(f"{n:>6} {trace_n:>10} {det_n:>14.6f}")

# (i) Pattern for trace
print(f"\n(i) [PATTERN - TRACE]")
print(f"    Traces: {traces}")
print(f"    Pattern: Trace of I_n = n (equals matrix size)")
print(f"    This is because trace sums diagonal elements, and identity has n ones on diagonal.")

# (ii) Pattern for determinant
print(f"\n(ii) [PATTERN - DETERMINANT]")
print(f"     Determinants: {[round(d, 6) for d in determinants]}")
print(f"     Pattern: det(I_n) = 1 for all n")
print(f"     This is because identity matrix represents no scaling/volume change.")

# (iii) General rule
print(f"\n(iii) [GENERAL RULE]")
print(f"      For any n×n identity matrix I_n:")
print(f"      • Trace(I_n) = n")
print(f"      • det(I_n) = 1")
print(f"      • These properties make identity matrix the multiplicative identity in linear algebra.")
```

## PROGRAM G — Evenly Spaced ndarray

```python
print("\n" + "=" * 60)
print("PROGRAM G — Evenly Spaced ndarray")
print("=" * 60)

# -----------------------------------------------------------
# Example G1 — The Scale Builder
# Context: Customer satisfaction score scale (0 to 10)
# Using linspace because we need exactly 11 points (0-10 inclusive)
# -----------------------------------------------------------
print("\n--- Example G1: Customer Satisfaction Scale ---")

# Using linspace: we need exactly 11 points with equal spacing
# Justification: We want specific count (11 points), not specific step size
satisfaction_scale = np.linspace(0, 10, 11)
print("Customer Satisfaction Scale (0-10, 11 points):")
print(satisfaction_scale)
print("[Used np.linspace because we need exactly 11 rating points]")

# Apply 5 real measurements from recent restaurant reviews
# Ratings from Google Reviews for local restaurants in Vizag
measurements = np.array([7.5, 8.2, 6.0, 9.1, 7.8])
print(f"\n5 Real Restaurant Ratings: {measurements}")

# Classify into zones
zones = []
for m in measurements:
    if m <= 4:
        zones.append('Poor')
    elif m <= 6:
        zones.append('Average')
    elif m <= 8:
        zones.append('Good')
    else:
        zones.append('Excellent')

print(f"\nClassifications:")
for m, z in zip(measurements, zones):
    print(f"  {m}: {z}")

# Count percentages
zone_counts = {'Poor': 0, 'Average': 0, 'Good': 0, 'Excellent': 0}
for z in zones:
    zone_counts[z] += 1

print(f"\nZone Distribution:")
for zone, count in zone_counts.items():
    pct = (count / len(measurements)) * 100
    print(f"  {zone}: {count} ({pct:.0f}%)")


# -----------------------------------------------------------
# Example G2 — The Schedule Generator
# Context: College bus timings from 6:00 AM to 10:00 PM every 30 minutes
# Using arange because we control the step (30 minutes)
# -----------------------------------------------------------
print("\n--- Example G2: College Bus Schedule ---")

# Convert time to minutes from midnight
# 6:00 AM = 360 minutes, 10:00 PM = 1320 minutes
# Step = 30 minutes
bus_times_minutes = np.arange(360, 1321, 30)  # 1321 to include 10:00 PM

print("College Bus Schedule (6:00 AM to 10:00 PM, every 30 min):")
print("[Used np.arange because we control the step size of 30 minutes]")

def minutes_to_time(m):
    hours = m // 60
    mins = m % 60
    ampm = "AM" if hours < 12 else "PM"
    hours = hours if hours <= 12 else hours - 12
    hours = 12 if hours == 0 else hours
    return f"{hours}:{mins:02d} {ampm}"

for i, t in enumerate(bus_times_minutes):
    print(f"  Slot {i+1}: {minutes_to_time(t)}")

# (i) How many total slots are generated?
total_slots = len(bus_times_minutes)
print(f"\n(i) Total bus slots: {total_slots}")

# (ii) Which slot number is closest to noon?
noon_minutes = 720
closest_to_noon_idx = np.argmin(np.abs(bus_times_minutes - noon_minutes))
closest_slot = closest_to_noon_idx + 1
print(f"\n(ii) Slot closest to noon (12:00 PM): Slot {closest_slot} ({minutes_to_time(bus_times_minutes[closest_to_noon_idx])})")

# (iii) If 3 slots are cancelled, how many remain and what is the new utilisation rate?
cancelled = 3
remaining = total_slots - cancelled
utilisation = (remaining / total_slots) * 100
print(f"\n(iii) After cancelling {cancelled} slots:")
print(f"       Remaining slots: {remaining}")
print(f"       Utilisation rate: {utilisation:.1f}%")


# -----------------------------------------------------------
# Example G3 — The Discovery
# Context: Compound interest comparison at 5%, 8%, and 12% over 20 years
# Find: At which year does 12% investment double the 5% investment?
# Using arange because we control step (1 year)
# -----------------------------------------------------------
print("\n--- Example G3: Compound Interest Discovery ---")

principal = 100000  # Rs 1 lakh
years = np.arange(0, 21, 1)  # 0 to 20 years, step 1

# [Used np.arange because we need yearly steps of exactly 1 year]
print("Compound Interest Growth (Principal: Rs 1,00,000)")
print("[Used np.arange because we control the step size of 1 year]")

rate_5 = principal * (1.05 ** years)
rate_8 = principal * (1.08 ** years)
rate_12 = principal * (1.12 ** years)

print(f"\n{'Year':>6} {'5%':>12} {'8%':>12} {'12%':>12} {'12%/5%':>10}")
print("-" * 58)

for i, y in enumerate(years):
    ratio = rate_12[i] / rate_5[i]
    print(f"{y:>6} {rate_5[i]:>12.0f} {rate_8[i]:>12.0f} {rate_12[i]:>12.0f} {ratio:>10.2f}x")

# Find threshold: when does 12% value become > 2x the 5% value?
ratio_array = rate_12 / rate_5
threshold_mask = ratio_array > 2
threshold_years = years[threshold_mask]

if len(threshold_years) > 0:
    first_threshold_year = threshold_years[0]
    print(f"\n(i) Threshold found: Year {first_threshold_year}")
    print(f"    At year {first_threshold_year}: 12% value = Rs {rate_12[first_threshold_year]:.0f}")
    print(f"    5% value = Rs {rate_5[first_threshold_year]:.0f}")
    print(f"    Ratio = {ratio_array[first_threshold_year]:.2f}x")
else:
    print(f"\n(i) 12% investment does not double the 5% investment within 20 years")

# (ii) One-line conclusion
print(f"\n(ii) [CONCLUSION]")
print(f"      # At {first_threshold_year} years, the 12% investment becomes more than double the 5% investment,")
print(f"      # demonstrating the power of compound interest rate differences over time.")

# (iii) Function justification
print(f"\n(iii) [FUNCTION JUSTIFICATION]")
print(f"      Used np.arange(0, 21, 1) because:")
print(f"      - We need yearly intervals (step = 1)")
print(f"      - np.linspace would give fractional years which don't make sense for annual compounding")
print(f"      - arange naturally generates integer year indices")

print("\n" + "=" * 60)
print("ASSIGNMENT COMPLETE — All Programs Executed Successfully")
print("=" * 60)
```
